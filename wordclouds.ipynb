# Wordclouds

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

## Load the data

```python
# load the data
df1 = pd.read_csv('data_raw/results_feralswine_05012019to11042019.csv')
df2 = pd.read_csv('data_raw/results_wildpigs_05012019to11042019.csv')
df3 = pd.read_csv('data_raw/results_wildhogs_05012019to11042019.csv')

df = pd.concat([df1, df2, df3], sort=True)

df = df.loc[df.relevant == 1., :]
df.shape
```

```python
df.head()
```

## Location Wordcloud

```python
text = " ".join(str(review) for review in df['user_loc'] if str(review) != 'nan' and
                                                            "Ä" not in str(review) and 
                                                            "á" not in str(review) and
                                                            "à" not  in str(review) and
                                                            "ü" not in str(review))
text = text.replace(' nan ', ' ')

stopwords = set(STOPWORDS)
stopwords.update(["thing", "one", "https", "new", "things"])

wordcloud = WordCloud(width=1600, height=800, max_words=100, collocations=True, stopwords=stopwords).generate(text)

plt.rcParams['figure.figsize'] = [6, 6]

%matplotlib notebook
plt.figure( figsize=(8, 4), facecolor='k')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()

wordcloud.to_file("location.png")
```
